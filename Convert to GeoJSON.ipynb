```python
def xl2geo(file):
    #read file into pandas and drop all NaN's
    temp_df = pd.read_excel(file)
    temp_df = temp_df.dropna()
    #convert string arrays into normal arrays
    temp_df['Location'] = temp_df['Location'].apply(lambda x: ast.literal_eval(x))
    features = []
    for i, row in temp_df.iterrows():
        props = dict()
        geom = geojson.Point(row.Location)
        props['Title'] = row['Title']
        props['Date'] = row['Date']
        props['Description'] = row['Description']
        props['Faction'] = row['Faction']
        props['Type'] = row['Type']
        f1 = geojson.Feature(geometry=geom, properties=props)
        features.append(f1)
    f_col = geojson.FeatureCollection(features)
    return f_col

```

```python
geoj = xl2geo('Wars of the Roses Test.xlsx')
with open('C:/Users/cwebb/Desktop/Wars-of-the-Roses-Map/events.geojson', 'w') as f:
    geojson.dump(geoj, f)
    f.close()
    
```
